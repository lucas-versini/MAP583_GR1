```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read file graphs/values.txt
header = []
mae = []
maeDEX = []
eps = []
epsDEX = []

with open('graphs/values.txt', 'r') as f:
    for _ in range(12):
        header.append(f.readline().strip())
        mae_list = f.readline().strip().split(',')
        mae.append([float(x) for x in mae_list])
        maeDEX_list = f.readline().strip().split(',')
        maeDEX.append([float(x) for x in maeDEX_list])
        eps_list = f.readline().strip().split(',')
        eps.append([float(x) for x in eps_list])
        epsDEX_list = f.readline().strip().split(',')
        epsDEX.append([float(x) for x in epsDEX_list])
        f.readline()
```

```python
# Paramètres des graphes,
plt.rc('figure', dpi = 400, figsize = (15,8))
plt.rc('font',size = 12, weight = 500)
plt.rc('lines', linewidth = 2)
colors = plt.cycler('color',['#e63946', '#fcd581', '#a1c349', '#3399cc', '#ff9ebb', '#C2BBF0', '#ccee66', '#f27f34', '#006699', '#6a853f'])
#[Rouge, jaune, vert, bleu, rose, violet, vert clair, orange, bleu foncé, vert foncé]
plt.rc('axes', facecolor='#FAFAFA', edgecolor='C3C3C3', axisbelow=True, grid = 'false', prop_cycle=colors)
```

```python
for i in range(6):
    plt.plot(mae[2*i], label=header[2*i].split(',')[0])
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.title('MAE for equal buckets')
plt.legend()
plt.show()
```

```python
for i in range(6):
    plt.plot(mae[2*i+1], label=header[2*i+1].split(',')[0])
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.title('MAE for balanced buckets')
plt.legend()
plt.show()
```

```python
for i in range(6):
    plt.plot(mae[2*i], label=header[2*i])
    plt.plot(mae[2*i+1], label=header[2*i+1], linestyle='dashed')
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.title('MAE for different bucketing methods')
plt.legend()
plt.show()
```
